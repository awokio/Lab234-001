```python
initial_data = {
'Symbol': ['A', 'B', 'C', 'D', 'F', 'G', 'H', 'I', 'J', 'K', 'L'],
'Origin': ['Other Asian', 'Black', 'Chinese', 'Cambodian', 'Filipino',
'Guamanian', 'Hispanic/Latin/Mexican', 'American Indian/Alaskan Native',
'Japanese', 'Korean', 'Laotian']
}
data = pd.DataFrame(initial_data)

origin_counts = data['Origin'].value_counts()

most_common_origin = origin_counts.idxmax()
print('Происхождение, которое чаще всего подвергается преступлениям:', most_common_origin)
```

```python
crime_counts_by_district = df['area_name'].value_counts()

sorted_districts = crime_counts_by_district.sort_values()

districts = sorted_districts.index
crime_counts = sorted_districts.values

plt.figure(figsize=(12, 8))
plt.barh(districts, crime_counts, color='skyblue')
plt.xlabel('Количество преступлений')
plt.ylabel('Районы')
plt.title('Количество преступлений в различных районах')
plt.show()
```

```python
top_crimes = df['crime_code_description'].value_counts().head(10)

plt.figure(figsize=(20, 12))
top_crimes.plot(kind='bar', rot=40)
plt.title('Топ-10 самых распространенных преступлений в Лос-Анджелесе')
plt.xlabel('Тип преступления')
plt.ylabel('Количество')
plt.tight_layout()

# Отображение графика
plt.show()
```
